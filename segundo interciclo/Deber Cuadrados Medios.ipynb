```python
def get_pos(digs):
    val1 =0
    val2 =0
    if digs%2 !=0:
        val1 = int(digs/2)
        val2 = int(digs/2)+1
    else:
        val1 = int(digs/2)
        val2 = int(digs/2)
    return val1,val2
```

```python
def calcular_num(iters, val, digs):
    x0_semilla = int(val)
    aum = get_pos(digs)
    print("ITERACIÓN", "\t Xn", "\t Xn*Xn", "\t\t Longitud","\t Ui","\t Rn")
    for i in range(iters):
        xn2= x0_semilla**2
        lon = len(str(xn2))
        ui = str(xn2)[int(lon/2)-aum[0]:int(lon/2)+aum[1]]
        rn = int(ui)/10**digs
        arreglorn.append(rn)
        #df=pd.DataFrame({"Xn":x0_semilla, "Xn*Xn":xn2 ,"Longitud":lon, "UI ":ui, "RN":rn})
        print("   ",i, "\t\t", x0_semilla,"\t",xn2, "\t", lon, "\t\t",ui, "\t", rn)
        x0_semilla=int(ui)
    print("   ")
    
iters = int(input("Iteraciones: "))
digs = int(input("Ingrese el digito: "))
for i in valores:
    print("****************************************************************")
    print("i: ", i)
    calcular_num(iters, i, digs)
```
